```python
# ==========================================================
# SIMULADOR IA: MAKEUP AGENDA (Maquillista & Memoria IA)
# ==========================================================

print("==================================================")
print("     ¡BIENVENIDO A AI MAKEUP AGENDA v3.0!        ")
print("  Gestiona tus modelos con la memoria de la IA    ")
print("==================================================")

# Estructura solicitada: Lista vacía al inicio (Memoria de la IA)
lista_modelos = []

while True:
    # Impresión del Menú Principal
    print("\n--- MENÚ DEL MAQUILLISTA ---")
    print("1. Registrar nueva Modelo y sugerir estilo")
    print("2. Ver historial de Modelos atendidas (Memoria IA)")
    print("3. Limpiar historial de trabajo")
    print("4. Cerrar Agenda Inteligente (Salir)")

    opcion = input("\nSelecciona una opción (1-4): ")

    # ------------------------------------------------------
    # OPCIÓN 1: REGISTRAR MODELO Y SUGERIR ESTILO
    # ------------------------------------------------------
    if opcion == "1":
        print("\n--- REGISTRO DE PACIENTE DE BELLEZA ---")
        nombre = input("¿Cuál es el nombre de la modelo?: ")

        # Guardamos el nombre en nuestra lista (Memoria de la IA)
        lista_modelos.append(nombre)

        print(f"¡Perfecto! '{nombre}' ha sido guardada en la memoria.")

        # Mini test de la IA para elegir el estilo de maquillaje
        print("\n¿Qué estilo prefiere la modelo?")
        print("- (g) Glam / Fiesta")
        print("- (n) Natural / Diario")
        estilo = input("Elige una letra (g / n): ")

        if estilo == "g":
            print(f"🤖 IA Sugiere: Para {nombre}, usa sombras brillantes y labios intensos.")
        elif estilo == "n":
            print(f"🤖 IA Sugiere: Para {nombre}, usa una 'skin tint' ligera y tonos durazno.")
        else:
            print("🤖 IA Nota: Estilo no reconocido, pero la modelo ya quedó registrada.")

    # ------------------------------------------------------
    # OPCIÓN 2: VER HISTORIAL DE MODELOS (Uso de la Lista)
    # ------------------------------------------------------
    elif opcion == "2":
        print("\n--- CONSULTANDO MEMORIA DE LA IA ---")

        # Comprobamos si la lista está vacía usando condicionales
        if lista_modelos == []:
            print("❌ La lista está vacía. Aún no has registrado a ninguna modelo hoy.")
        else:
            print("✅ Modelos atendidas en esta sesión:")
            print(lista_modelos)
            # Contamos cuántas modelos van usando la función len() incorporada de Python
            print(f"Total de trabajos realizados: {len(lista_modelos)}")

    # ------------------------------------------------------
    # OPCIÓN 3: LIMPIAR HISTORIAL (Vaciar la Lista)
    # ------------------------------------------------------
    elif opcion == "3":
        print("\n--- BORRANDO DATOS ---")
        if lista_modelos == []:
            print("La lista ya estaba vacía, no hay nada que borrar.")
        else:
            lista_modelos.clear() # Vacía por completo la lista
            print("♻️ Memoria de la IA formateada. La lista vuelve a estar vacía [].")

    # ------------------------------------------------------
    # OPCIÓN 4: SALIR DEL PROGRAMA
    # ------------------------------------------------------
    elif opcion == "4":
        print("\nCerrando IA Makeup Agenda... ¡Que tengas un excelente día de trabajo!")
        break  # Rompe el ciclo infinito while True

    # CONDICIONAL POR SI ELIGEN UNA OPCIÓN INEXISTENTE
    else:
        print("❌ Opción inválida. Elige un número del 1 al 4.")

    print("\n--------------------------------------------------")

print("\n==================================================")
print("     Agenda fuera de línea. ¡Hasta pronto!        ")
print("==================================================")
print("Programa realizado por Arath Valenzuela 0148 NL57")
```
